```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("../data/contact-area-distributed.csv")
data.head(2)
y = data["resistance [Ohm]"]
x = data["n_pieces"]
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(x, y, '-ok', label=r'$\varepsilon = 0.25$')
ax.set_xlabel(r"$n_{pieces}$", fontsize=20)
ax.set_ylabel(r"R [$\Omega$]", fontsize=20)
ax.set_title(r"$\kappa_0$ = 1 S/m", fontsize=20)
ax.legend(fontsize=20)
plt.savefig("../figures/contact-area-distributed.png", dpi=500)
plt.show()
```

```python
df = pd.read_csv("../data/contact-area-varied-slits.csv")
df.head(2)
slit_pt25 = df.loc[df["slit pos [y*]"] == 0.25]
slit_pt5 = df.loc[df["slit pos [y*]"] == 0.5]
slit_pt75 = df.loc[df["slit pos [y*]"] == 0.75]
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(slit_pt25['eps'], slit_pt25['resistance [Ohm]']/100, 'ok', label=r'$y^* = 0.25$')
ax.plot(slit_pt5['eps'], slit_pt5['resistance [Ohm]']/100, label=r'$y^* = 0.50$')
ax.plot(slit_pt75['eps'], slit_pt75['resistance [Ohm]']/100, label=r'$y^* = 0.75$')
ax.set_ylabel(r'R[$\Omega$]', fontsize=20)
ax.set_xlabel(r"$\varepsilon$", fontsize=20)
ax.set_title(r"$W = 100$, $H = 1$, $w = 50$, $h = 0.1$, $\kappa_0 = 1$ S/m", fontsize=20)
ax.legend(fontsize=20)
plt.savefig("../figures/contact-area-varied-slits.png", dpi=500)
```
