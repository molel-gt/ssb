```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
figsize = (5, 4.5)
labelsize = 12
labelpad = 10
fontsize = 12
```

```python
def Î´(i: int, j: int) -> int:
    if i != j:
        return 0
    return 1
```

```python
def basis_function(i, x, x_arr, h=0.1):
    return 1/h * (x - x_arr[i-1]) * (x_arr[i-1] <= x <= x_arr[i]) + 1/h * (x_arr[i+1] - x) * (x_arr[i] < x <= x_arr[i+1])
```

```python
interval = np.linspace(0, 1, num=11)
h = 0.1
fun0 = lambda x: basis_function(4, x, x_arr=interval, h=h)
fun1 = lambda x: basis_function(5, x, x_arr=interval, h=h)
fun2 = lambda x: basis_function(6, x, x_arr=interval, h=h)
```

```python
interval2 = np.linspace(0, 1, num=101)
y0 = [fun0(v) for v in interval2]
y1 = [fun1(v) for v in interval2]
y2 = [fun2(v) for v in interval2]
```

```python
fig, ax = plt.subplots(figsize=figsize)
ax.plot(interval2, y0, label='j-1', color='red');
ax.plot(interval2, y1, label='j', color='blue');
ax.plot(interval2, y2, label='j+1', color='green');
ax.set_box_aspect(1);
ax.set_xlabel('x', fontsize=fontsize);
ax.minorticks_on();
ax.tick_params(axis='both', which='major', labelsize=labelsize);
# ax.set_ylim([0, 1]);
plt.tight_layout();
```
