```python
fig, ax = plt.subplots(1, 4, figsize=(20, 4.5))
ax[0].plot(df['Cu foil'][2:], -df['Unnamed: 1'][2:])
ax[0].set_xlabel(r'$Z_{Re}$($\Omega$)');
ax[0].set_ylabel(r'-$Z_{Im}$($\Omega$)');
ax[0].set_title('Pristine');

ax[1].plot(df['Cu foil.1'][2:], -df['Unnamed: 6'][2:])
ax[1].set_xlabel(r'$Z_{Re}$($\Omega$)');
ax[1].set_ylabel(r'-$Z_{Im}$($\Omega$)');
ax[1].set_title('1st cycle');

ax[2].plot(df['Cu foil.2'][2:], -df['Unnamed: 11'][2:])
ax[2].set_xlabel(r'$Z_{Re}$($\Omega$)');
ax[2].set_ylabel(r'-$Z_{Im}$($\Omega$)');
ax[2].set_title('5th cycle');

ax[3].plot(df['Cu foil.3'][2:], -df['Unnamed: 16'][2:])
ax[3].set_xlabel(r'$Z_{Re}$($\Omega$)');
ax[3].set_ylabel(r'-$Z_{Im}$($\Omega$)');
ax[3].set_title('20th cycle');
plt.tight_layout();
plt.savefig('figures/3dcc/foil-cu-eis.png');
```
